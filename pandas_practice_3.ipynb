## Q8

국가의 '% Renewable' 값이 상위 15 개 국가의 중앙값 이상이면 1, 국가의 재생 가능 비율 값이 중앙값 미만이면 0으로 새 열을 만듭니다.

````python
mid_renew = final['% Renewable'].median()
````

````python
def highrenew(renew):
    if renew>=mid_renew:
        return 1
    else:
        return 0
    
final['HighRenew'] = final['% Renewable'].apply(lambda x: highrenew(x))
````

````python
final['HighRenew']
````

````python
ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
````

````python
cont = pd.DataFrame.from_dict(ContinentDict, orient='index', columns=['Continent'])
````

````python
q9 = final.join(cont)
````

````python
q9
````

````python
q9 =q9[['Continent', 'Capita']]
q9
````

````python
q9.info()
````

````python
q9['Capita'] = pd.to_numeric(q9['Capita'])
````

````python
q9.info()
````

````python
group=q9.groupby(q9['Continent'])
````

````python
group_size = group.count().rename(columns={'Capita': 'size'})
group_sum = group.sum().rename(columns={'Capita': 'sum'})
group_mean = group.mean().rename(columns={'Capita': 'mean'})
group_std = group.std().rename(columns={'Capita': 'std'})
````

````python
group_sum
````

````python
continent = group_size.join(group_sum).join(group_mean).join(group_std)
````

````python
continent
````
